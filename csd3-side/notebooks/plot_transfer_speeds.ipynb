```python
import matplotlib.pyplot as plt
import os, sys
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
logfile = os.path.expanduser('~/five_runs.csv')
```

```python
title_name = logfile.split("/")[-1].split(".")[0]
title_name
```

```python
df = pd.read_csv(logfile)
df['UPLOAD_START'] = pd.to_datetime(df['UPLOAD_START'])
df['UPLOAD_END'] = pd.to_datetime(df['UPLOAD_END'])
df
```

```python
df['SPEED'] = df['FILE_SIZE'] / df['UPLOAD_TIME']
df.dtypes
```

```python
all_start = df['UPLOAD_START'].min()
all_end = df['UPLOAD_END'].max()
all_start, all_end
```

```python
time_axis = pd.date_range(all_start, all_end, freq='1s')
time_axis
```

```python
sum_speeds = []
concurrency = []
for i in range(len(time_axis)):
    sum_speeds.append(
        df[(df['UPLOAD_START'] <= time_axis[i]) & (df['UPLOAD_END'] >= time_axis[i])]['SPEED'].sum()
    )
    concurrency.append(len(df[(df['UPLOAD_START'] <= time_axis[i]) & (df['UPLOAD_END'] >= time_axis[i])]))
sum_speeds = np.array(sum_speeds)
sum_speeds = (sum_speeds * 8) / 1024**3  # convert to Gbps
sum_speeds
concurrency = np.array(concurrency)
```

```python
per_uploading_process = np.divide(sum_speeds,concurrency, out=np.zeros_like(sum_speeds), where=concurrency!=0)
```

```python
print(len(time_axis), len(sum_speeds), len(concurrency), len(per_uploading_process))
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
plt.plot(time_axis, sum_speeds, label='Total Upload Speed', color='blue')
plt.plot(time_axis, per_uploading_process, label='Mean Individual Upload Speed', color='red', alpha=0.5)
plt.fill_between(time_axis, sum_speeds, color='blue', alpha=0.1)
plt.title(f"Total Upload Speed over Time\n{title_name}", fontsize=10)
plt.xlabel("Time", fontsize=10)
plt.ylabel("Upload Speed (Gbps)", fontsize=10)
# plt.xlim(time_axis[0], time_axis[-180])
plt.xticks(rotation=45, fontsize=10)
plt.grid()
plt.legend()
plt.tight_layout()
# plt.show()
plt.savefig(os.path.expanduser(f"~/{title_name}_total_upload_speed.png"), dpi=300)
```

```python
# sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
plt.plot(time_axis, concurrency, label='Uploads', color='green')
plt.fill_between(time_axis, concurrency, color='green', alpha=0.1)
plt.title(f"Concurrent Uploads over Time\n{title_name}", fontsize=10)
plt.xlabel("Time", fontsize=10)
plt.ylabel("Concurrent Uploads", fontsize=10)
# plt.xlim(time_axis[0], time_axis[-180])
plt.xticks(rotation=45, fontsize=10)
plt.grid()
# plt.legend()
plt.tight_layout()
# plt.show()
plt.savefig(os.path.expanduser(f"~/{title_name}_concurrent_uploads.png"), dpi=300)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(time_axis, per_uploading_process, label='Total Upload Speed per Uploading Process', color='blue')
plt.fill_between(time_axis, per_uploading_process, color='blue', alpha=0.1)
plt.title(f"Total Upload Speed per Uploading Process", fontsize=10)
plt.xlabel("Time", fontsize=10)
plt.ylabel("Upload Speed (Gbps)", fontsize=10)
# plt.xlim(time_axis[0], time_axis[-1])
plt.xticks(rotation=45, fontsize=10)
plt.grid()
# plt.legend()
plt.tight_layout()
# plt.show()
plt.savefig(os.path.expanduser(f"~/{title_name}_total_upload_per_proc_speed.png"), dpi=300)
```
