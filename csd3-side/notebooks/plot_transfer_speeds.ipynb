```python
import matplotlib.pyplot as plt
import os, sys
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
logfile = os.path.expanduser('~/butler_full_20221201-data-u-lsst-backup3.csv')
```

```python
title_name = logfile.split("/")[-1].split(".")[0]
title_name
```

```python
df = pd.read_csv(logfile)
df['UPLOAD_START'] = pd.to_datetime(df['UPLOAD_START'])
df['UPLOAD_END'] = pd.to_datetime(df['UPLOAD_END'])
df
```

```python
df['SPEED'] = df['FILE_SIZE'] / df['UPLOAD_TIME']
df.dtypes
```

```python
all_start = df['UPLOAD_START'].min()
all_end = df['UPLOAD_END'].max()
all_start, all_end
```

```python
interval = 60 # seconds
```

```python
time_axis = pd.date_range(all_start, all_end, freq=f'{interval}s')
time_axis
```

```python
sum_speeds = []
concurrency = []
for i in range(len(time_axis)):
    sum_speeds.append(
        df[(df['UPLOAD_START'] <= time_axis[i]) & (df['UPLOAD_END'] >= time_axis[i])]['SPEED'].sum()
    )
    concurrency.append(len(df[(df['UPLOAD_START'] <= time_axis[i]) & (df['UPLOAD_END'] >= time_axis[i])]))
sum_speeds = np.array(sum_speeds)
sum_speeds = (sum_speeds * 8) / 1024**3  # convert to Gbps
sum_speeds
concurrency = np.array(concurrency)
```

```python
per_uploading_process = np.divide(sum_speeds,concurrency, out=np.zeros_like(sum_speeds), where=concurrency!=0)
```

```python
print(len(time_axis), len(sum_speeds), len(concurrency), len(per_uploading_process))
```

```python
df = pd.DataFrame.from_dict({
    'time': time_axis,
    'total_speeds': sum_speeds,
    'process_speeds': sum_speeds,
    'concurrency': concurrency,
})
df = df[df['total_speeds'] > 1]
```

```python
df['time']
```

```python
# sns.set(style="whitegrid")
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, sharey=True)
# plt.xticks(rotation=45, fontsize=10)
f.suptitle(f"Total Upload Speed over Time\n{title_name}", fontsize=10)
plt.xlabel("Time", fontsize=10)
for ax in (ax1, ax2, ax3, ax4):
    ax.plot(df['time'], df['total_speeds'], label='Total Upload Speed', color='blue')
    ax.plot(df['time'], df['process_speeds'], label='Mean Individual Upload Speed', color='red', alpha=0.5)
    ax.fill_between(time_axis, sum_speeds, color='blue', alpha=0.1)
    ax.tick_params(rotation=45, labelsize=8)
    # ax.tick_params(labelleft='off')
    # ax.grid()

ax1.set_xlim(time_axis[0], time_axis[40])
ax1.spines['right'].set_visible(False)
ax1.yaxis.tick_left()
ax2.set_xlim(time_axis[1140], time_axis[1190])
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax3.set_xlim(time_axis[2700], time_axis[2750])
ax3.spines['right'].set_visible(False)
ax3.spines['left'].set_visible(False)

ax4.set_xlim(time_axis[3600], time_axis[3643])
ax4.spines['left'].set_visible(False)
ax4.yaxis.tick_right()

ax1.set_ylabel("Upload Speed (Gbps)", fontsize=10)



# plt.legend()
# plt.tight_layout()
# plt.show()
plt.savefig(os.path.expanduser(f"~/{title_name}_total_upload_speed.png"), dpi=300)
```

```python
# sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['concurrency'], label='Uploads', color='green')
plt.fill_between(df['time'], df['concurrency'], color='green', alpha=0.1)
plt.title(f"Concurrent Uploads over Time\n{title_name}", fontsize=10)
plt.xlabel("Time", fontsize=10)
plt.ylabel("Concurrent Uploads", fontsize=10)
# plt.xlim(time_axis[0], time_axis[-180])
plt.xticks(rotation=45, fontsize=10)
plt.grid()
# plt.legend()
plt.tight_layout()
# plt.show()
plt.savefig(os.path.expanduser(f"~/{title_name}_concurrent_uploads.png"), dpi=300)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['process_speeds'], label='Total Upload Speed per Uploading Process', color='blue')
plt.fill_between(df['time'], df['process_speeds'], color='blue', alpha=0.1)
plt.title(f"Total Upload Speed per Uploading Process", fontsize=10)
plt.xlabel("Time", fontsize=10)
plt.ylabel("Upload Speed (Gbps)", fontsize=10)
# plt.xlim(time_axis[0], time_axis[-1])
plt.xticks(rotation=45, fontsize=10)
plt.grid()
# plt.legend()
plt.tight_layout()
# plt.show()
plt.savefig(os.path.expanduser(f"~/{title_name}_total_upload_per_proc_speed.png"), dpi=300)
```
