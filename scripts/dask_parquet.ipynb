```python
from dask import dataframe as dd
```

```python
import numpy as np
```

```python
import os
```

```python
data = np.random.rand(1000000, 100)
data
```

```python
ddf = dd.from_array(data, chunksize=1000, columns=[f'col_{i}' for i in range(100)])
```

```python
ddf
```

```python
ddf.visualize()
```

```python
ddf.to_parquet(os.path.expanduser('~') + '/data.parquet')
```
